## Cleaning

```python
def cleaning(text):
  text = re.sub(r'[^a-zA-Z\s]', '', text).strip()
  return text

df['Cleaning'] = df['Artikel'].apply(cleaning)
df['Cleaning']
```

Script tersebut berisi fungsi dan langkah-langkah untuk melakukan pembersihan teks dalam sebuah DataFrame (`df`). Berikut adalah penjelasan untuk setiap komponen:

1. **Fungsi `cleaning(text)`**:
   - `text = re.sub(r'[^a-zA-Z\s]', '', text).strip()`: Memanfaatkan modul ekspresi reguler (`re`) untuk menghapus semua karakter selain huruf (baik huruf besar maupun huruf kecil) dan spasi. Metode `strip()` digunakan untuk menghapus spasi di awal dan akhir teks.
   - `return text`: Mengembalikan teks yang telah dibersihkan.

2. **Penggunaan fungsi pada DataFrame**:
   - `df['Artikel'].apply(cleaning)`: Menggunakan fungsi `cleaning` untuk membersihkan teks pada kolom 'Artikel' dari DataFrame `df`.
   - `df['Cleaning']`: Membuat kolom baru ('Cleaning') dalam DataFrame yang berisi teks yang telah dibersihkan.

Secara keseluruhan, skrip ini berfungsi untuk membersihkan teks dalam kolom 'Artikel' dari DataFrame dan menyimpan hasilnya dalam kolom baru dengan nama 'Cleaning'. Teks yang telah dibersihkan ini dapat digunakan untuk analisis atau langkah-langkah pemrosesan lanjutan.

## Tokenizing

```python
def tokenizer(text):
  text = text.lower()
  return word_tokenize(text)

df['Tokenizing'] = df['Cleaning'].apply(tokenizer)
df['Tokenizing']
```

Kode ini bertujuan untuk melakukan tokenisasi pada teks yang telah dibersihkan dalam kolom 'Cleaning' dari DataFrame dan menyimpan hasilnya dalam kolom baru dengan nama 'Tokenizing'. Proses ini bermanfaat untuk memecah teks menjadi unit-unit terkecil (token) seperti kata-kata, sehingga mempermudah analisis dan pemrosesan lanjutan.

## Stopwords

```python
corpus = stopwords.words('indonesian')

def stopwordText(words):
 return [word for word in words if word not in corpus]

df['Stopword Removal'] = df['Tokenizing'].apply(stopwordText)

# Gabungkan kembali token menjadi kalimat utuh
df['Full Text'] = df['Stopword Removal'].apply(lambda x: ' '.join(x))
df['Full Text']
```

 Kode ini melakukan penghapusan kata-kata penghenti dari teks yang telah di-tokenisasi, dan kemudian menggabungkan kembali token-token tersebut menjadi kalimat utuh. Proses ini bertujuan untuk membersihkan teks dari kata-kata yang umum dan tidak memberikan kontribusi signifikan pada analisis teks atau pemrosesan lanjutan.

## TF-IDF

```python
def tfidf(dokumen, category):
  vectorizer = TfidfVectorizer()
  x = vectorizer.fit_transform(dokumen).toarray()
  terms = vectorizer.get_feature_names_out()

  final_tfidf = pd.DataFrame(x, columns=terms)
  final_tfidf.insert(0, 'Dokumen', dokumen)
  final_tfidf.insert(len(final_tfidf.columns),'Category', category)

  return (vectorizer, final_tfidf)

tfidf_vectorizer, final_tfidf = tfidf(df['Full Text'], df['Category'])
final_tfidf
```

Pendekatan Term Frequency-Inverse Document Frequency (TF-IDF) untuk mengonversi teks dalam suatu DataFrame menjadi representasi vektor TF-IDF.

## Inisialisasi x dan y

Kode `X = final_tfidf.drop(['Dokumen', 'Category'], axis=1)` memiliki fungsi untuk membuat variabel `X` yang merupakan subset dari DataFrame `final_tfidf`. Secara khusus, variabel `X` akan berisi kolom-kolom dari `final_tfidf` kecuali kolom 'Dokumen' dan 'Category'. Dengan menggunakan metode `drop` pada DataFrame, kita menghapus kolom-kolom yang tidak diperlukan.

Jadi, `X` akan berisi representasi fitur dari DataFrame `final_tfidf` yang dapat digunakan sebagai input untuk pemodelan atau analisis lanjutan.
```python
X = final_tfidf.drop(['Dokumen', 'Category'], axis=1)
X
```

```python
y = df['Category']
y
```

## Split Data


Tujuan dari membagi data menjadi dua set yang terpisah, yaitu set pelatihan (training set) dan set pengujian (test set), adalah untuk mengukur sejauh mana model atau algoritma yang dikembangkan dapat melakukan generalisasi pada data yang tidak pernah dilihat sebelumnya

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Pemodelan LDA

### Pemodelan Parameter Optimal pada LDA

```python
def find_best_lda(Xtrain, Xtest, n_components, alpha, beta):
  looping = 1
  best = {'k' : 0, 'alpha' : 0, 'beta' : 0, 'accuracy' : 0, 'model': '', 'lda' : '', 'lda_Xtrain' : '', 'lda_Xtest' : ''}
  history = pd.DataFrame(columns=["Pengujian Ke", "K", "Alpha", "Beta", "Accuracy"])

  # Menambahkan tqdm pada loop terluar
  for k in n_components:
    for a in alpha:
      for b in beta:
        lda = LatentDirichletAllocation(n_components=k, doc_topic_prior=a, topic_word_prior=b)
        lda_Xtrain = lda.fit_transform(Xtrain)
        lda_Xtest = lda.transform(Xtest)

        # Membuat model Naive Bayes
        model = GaussianNB()

        # Melatih model pada data pelatihan
        model.fit(lda_Xtrain, y_train)

        # Melakukan prediksi pada data pengujian
        y_pred = model.predict(lda_Xtest)

        # Menghitung akurasi
        accuracy = accuracy_score(y_test, y_pred)
        print(f"Jumlah Topik: {k}, Alpha: {a}, Beta: {b}, Accuracy: {accuracy}")

        if accuracy > best['accuracy']:
          best['accuracy'] = accuracy
          best['k'] = k
          best['alpha'] = a
          best['beta'] = b
          best['model'] = model
          best['lda'] = lda
          best['lda_Xtrain'] = lda_Xtrain
          best['lda_Xtest'] = lda_Xtest

        history.loc[len(history)] = [f"Pengujian Ke- {looping}", k, a, b, accuracy]
        looping += 1

  return (best, history)

k = [3, 4, 5]
alpha = [0.3, 0.4]
beta = [0.1, 0.2]
best_param, history = find_best_lda(X_train, X_test, k, alpha, beta)
```

### LDA

```python
lda = best_param['lda']
lda_x_train = best_param['lda_Xtrain']
lda_x_test = best_param['lda_Xtest']
```

### Hasil Reduksi Dimensi

```python
topik_columns = [f"Topik {i}" for i in range(1, best_param['k']+1)]
dokumen = final_tfidf['Dokumen']
output_proporsi_TD = pd.DataFrame(lda_x_train, columns=topik_columns)
output_proporsi_TD.insert(0,'Dokumen', dokumen)
output_proporsi_TD.insert(len(output_proporsi_TD.columns),'Category', final_tfidf['Category'])
output_proporsi_TD
```

```python
output_proporsi_TD.to_csv('reduksi dimensi.csv', index=False)
```

### Proporsi Kata Setiap Topik

```python
# Output distribusi kata pada topik
distribusi_kata_topik = pd.DataFrame(lda.components_)
distribusi_kata_topik
```

### Pemodelan dengan Naive Bayes

```python
# Membuat model Naive Bayes
model = best_param['model']

# Melakukan prediksi pada data pengujian
y_pred = model.predict(lda_x_test)

# Menghitung akurasi
accuracy = accuracy_score(y_test, y_pred)
print("Akurasi:", accuracy)

# Menampilkan laporan klasifikasi
print("Laporan Klasifikasi:")
print(classification_report(y_test, y_pred))

# Menampilkan matriks kebingungan
confusion = confusion_matrix(y_test, y_pred)
print("Confusion Matriks:")
print(confusion)
```
